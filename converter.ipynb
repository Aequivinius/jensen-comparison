```python
import re
journal_re = re.compile('^\d+\. (.+)\.$')
year_re = re.compile('((1|2)\d\d\d)')
pmid_re = re.compile('^PMID: (\d+).*')
f =  open('pubmed_oligospermia.txt','r')
g = open('oligospermia.tsv','w')
pmid = None
author = None
title = None
text = None
for line in f:
    if len(line) > 1:
        journal_title = journal_re.search(line)
        if journal_title:
            if all(v is not None for v in [pmid, author, journal, year, title, text]):
                g.write('\t'.join([pmid, author, journal, year, title, text]) + '\n')
            journal = journal_title.group(1).strip()
            year = year_re.search(journal).group(1)
            pmid = None
            author = None
            title = None
            text = None
            continue
        if title == None:
            title = line.strip()
            print(title)
            continue
        if author == None:
            author = line.strip()
            continue
            
        if text == None:
            text = line.strip()
            continue
        
        pmid_result = pmid_re.search(line)
        if text and not pmid_result:
            text += ' '
            text += line.strip()
            continue
        if pmid_result:
            pmid = pmid_result.group(1)
            

# output    
# id author journal year title text
```

```python
import pandas

def prefix(string):
    return 'DOC:' + str(string)

def pubtator_to_jensen(input_path):
    input_data = pandas.read_csv(input_path, delimiter='|', 
                                 encoding='utf-8', header=None,
                                 names = ["PMID","type","text"])
    
    titles = input_data[input_data['type']=='t']
    titles.rename(columns={"text": "title"},inplace=True)
    titles.drop(columns=['type'],inplace=True)
    
    articles = input_data[input_data['type']=='a']
    articles.drop(columns=['type'],inplace=True)

    articles = articles.set_index('PMID').join(titles.set_index('PMID'))
    
    articles['author'] = 'Albus Dumbledore'
    articles['journal'] = 'Annals of All'
    articles['year'] = '1990'
    
    articles['index'] = articles.index
    articles['PMID'] = articles['index'].apply(prefix)
    
    articles = articles[['PMID','author','journal','year','title','text']]
    print(articles.head(15))
    articles.to_csv('output.tsv',sep='\t',index = False, header = False, encoding = 'utf-8')
    
    
pubtator_to_jensen('medline17n0842.pubtator')
```
